```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as skl
from sklearn import linear_model
```

```python
#importing the dataset
data=pd.read_csv('houseprice.csv')
#data cleaning
data.head()

```

```python
data.drop(['waterfront',	'view' ,'street','city','statezip','country'],axis=1)
```

```python
import seaborn as sns
df=data
sns.scatterplot(x='sqft_lot',y='price',hue='condition', data=df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['sqft_lot']],df[['price']])
```

```python
import numpy as np
import pandas as pd

# Create a DataFrame with the feature name 'sqft_lot'
prediction_input = pd.DataFrame({'sqft_lot': [50000]})

reg.predict(prediction_input)
```
